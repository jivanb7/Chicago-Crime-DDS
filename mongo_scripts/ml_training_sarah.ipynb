```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import (classification_report, ConfusionMatrixDisplay)

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('crime_aggregated.csv')
df.head()
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df_corr = df.copy().drop(columns=['primary_type'])
corr = df_corr.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Predictor Correlation Matrix")
plt.show()
```

Classification to predict crime type using Random Forest

```python
X = df.drop(columns="primary_type").copy()
X["night"] = (X["hour"] >= 20).astype(int)
X["weekend"] = (X["dow"] >= 6).astype(int)

le = LabelEncoder()
y  = le.fit_transform(df["primary_type"])
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y)
```

```python
model = RandomForestClassifier(n_estimators=200, max_depth=10, n_jobs=-1)
model.fit(X_train, y_train)
preds = model.predict(X_val)
```

```python
labels = np.unique(np.concatenate([y_val, preds]))
print(classification_report(y_val, preds, labels=labels, target_names=le.classes_))
```

Look at accuracy and feature importance

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 8))
ConfusionMatrixDisplay.from_predictions(y_val, preds, labels=labels, display_labels=le.classes_[labels], ax=axes[0], xticks_rotation=90, colorbar=False)
axes[0].set_title("Confusion Matrix")

feat_imp = pd.Series(model.feature_importances_, index=X.columns).sort_values()
feat_imp.plot(kind="barh", ax=axes[1])
axes[1].set_title("Feature Importance")

plt.tight_layout()
plt.show()
```
